```python
import pandas as pd
from autoscraper import AutoScraper
```

```python
url = "https://yoshops.com/products"
data = ['https://yoshops.com/products/reliance-jio-phone1', 'Reliance Jio Phone','https://isteam.wsimg.com/ip/a8efe83b-6857-477d-9d0f-f13ca0229a20/ols/1354_original/:/rs=w:600,h:600']
scrap = AutoScraper()
result = scrap.build(url, data)
print(result)
```

```python
scrap.get_result_similar(url, grouped=True)
```

```python
scrap.set_rule_aliases({'rule_o2ii': 'product_link','rule_36wm': 'Title' , 'rule_amkd' :"img_link"})
scrap.save('yoshops_products')
```

```python
page = [0]*12
page[0]=scrap.get_result_similar('https://yoshops.com/products',group_by_alias=True)
for i in range(1,12):
    page[i]=scrap.get_result_similar('https://yoshops.com/products?page='+str(i),group_by_alias=True)
```

```python
print(len(page))
```

```python
df = [0]*12
for i in range(12):
    df[i]=pd.DataFrame({"product-link":page[i]['product_link'],
                      "product_title": page[i]['Title'],
                      "product_image_link": page[i]['img_link']})
```

```python
df[0].head()
```

```python
final_df = pd.concat([df[0], df[1],df[2],df[3],df[4],df[5],df[6],df[7],df[8],df[9],df[10],df[11]], axis=0)
```
