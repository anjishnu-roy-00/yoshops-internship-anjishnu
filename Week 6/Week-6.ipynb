```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as num
```

1. Generating Payslips For November And December

```python
#Importing the Data
pay_nov = pd.read_excel('./Kalyani_Balance_Sheet_November_2022.xlsx')
pay_dec = pd.read_excel('./Kalyani_Balance_Sheet_December_2022.xlsx')
```

```python
#Function to generate the payslips
name_block = "Name of intern"
Department = 'Department'
foa = 'Fee of Aurangabad'
foc = 'Fee of Chennai'
def payslip(n, d, a, c):
    print('*' * 58)
    print('\t\t\tYOSHOPS\t\t')
    print('\t\tDATA SCIENCE INTERNSHIP PAYSIP')
    print('=' * 58)
    print('Branch Team Leader\t\t\t   Kalyani Ghale')
    print('=' * 58)
    print(f"{name_block.upper()}\t\t\t\t{n}")
    print(f"{Department.upper()}\t\t\t\t{d}")
    print(f"{foa.upper()}\t\t\t\t{a}")
    print(f"{foc.upper()}\t\t\t\t\t{c}")
    print('=' * 58)
    print(f'SUBTOTAL\t\t\t\t\t{a + c}')
    print('\n')
```

```python
# Let's user select b/w November or December Playslips
a = int(input('Enter a number press 1 for November and 2 for December :'))
if a == 1:
    print("Payslip for November Interns :")
    print()
    df = df = pay_nov.iloc[:, 0:6]
    for i in df.values:
            payslip(i[0], i[2], i[5], i[4])
elif a == 2:
    print("Payslip for December Interns :")
    print()
    df = df = pay_dec.iloc[:, 0:6]
    for i in df.values:
            payslip(i[0], i[2], i[5], i[4])
else:
    print('Invalid Selection')
print()
```

2. Generating Cash Flow Report and Balance Sheet for Chennai/Aurangabad(we use November Data)<br>
cash outflow/liabilities = 200/head (marketing+operational - 100 rs each)

```python
df = pay_nov.iloc[:, 0:6]
df.head()
```

```python
#No. of students
df.shape[0]
```

```python
ci = "Cash Inflow : "
co = "Cash Outflow : "
def gen_cash_flow(n,d):
    print('*' * 58)
    print('\t\t\tYOSHOPS\t\t')
    print('=' * 58)
    print('\t\tCASH FLOW STATEMENT')
    print('=' * 58)
    print(f"{ci.upper()}\t\t\t\t{n}")
    print(f"{co.upper()}\t\t\t\t{d}")
    print('=' * 58)
    print(f'TOTAL REVENUE\t\t\t\t\t{n - d}')
    print('\n')
```

```python
out = df.shape[0] * 200
a = input('Enter a number press a for chennai and b for Aurangabad :')
if a == 'a':
    sum = df['Chennai'].sum()
    print(f'\nTotal Cash Flow via Chennai:')
    gen_cash_flow(sum,out)
elif a == 'b':
    sum = df['Aurangabad'].sum()
    print(f'Total Cash Flow via Aurangabad:')
    gen_cash_flow(sum,out)
else:
    print('Invalid Selection')
print()
```

3. In Balance sheet we use Assets and Liabilities

```python
fe = "Fees : "
ma = "Marketing : "
op = "Operations : "
def gen_bal_sheet(m,n,o):
    print('*' * 58)
    print('\t\t\tYOSHOPS\t\t\t')
    print('=' * 58)
    print('\t\t\tBALANCE SHEET')
    print('=' * 58)
    print("ASSETS :")
    print('=' * 58)
    print(f"{fe.upper()}\t\t\t\t{m}")
    print("LIABILITIES:")
    print('=' * 58)
    print(f"{ma.upper()}\t\t\t{n}")
    print(f"{op.upper()}\t\t\t{o}")
    print('=' * 58)
    print(f'TOTAL REVENUE\t\t\t{m - (n + o)}')
    print('\n')
```

```python
out = df.shape[0] * 100
a = input('Enter a number press a for chennai and b for Aurangabad :')
if a == 'a':
    sum = df['Chennai'].sum()
    print(f'\nBALANCE SHEET - Chennai:')
    gen_bal_sheet(sum,out,out)
elif a == 'b':
    sum = df['Aurangabad'].sum()
    print(f'BALANCE SHEET - Aurangabad:')
    gen_bal_sheet(sum,out,out)
else:
    print('Invalid Selection')
print()
```

4. Stake Holder profit report

```python
sa = df['Aurangabad'].sum()
sc = df['Chennai'].sum()
labels = 'Aurangabad', 'Chennai'
sizes = [sa, sc]
explode = (0, 0.1)
```

```python
fig1, ax1 = plt.subplots()
colors = ['#F66F8D', '#FFB643']
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, colors=colors)
plt.title('Stake holder profit ratio in different cities', fontsize=14)
ax1.axis('equal')
plt.show(); 
```
